```python
import pandas as pd

import numpy as np

import scipy

import matplotlib.pyplot as plt

import matplotlib.dates as mdates

import datetime as dt
```

```python
# Input File paths

file = '../Data/combined_df_US.xlsx'
```

```python
US_df = pd.read_excel(file)

US_df.tail()
```

```python
# Apply Datetime format to date field

US_df['date'] = pd.to_datetime(US_df['date'])
```

```python
#Create fig and axis objects with subplots()

fig,ax = plt.subplots()

ax.plot(US_df.date, US_df.cases, color = "red", marker = "o")

plt.axvline(dt.datetime(2020, 3, 19), color = 'orangered', linestyle = '--')

plt.text(dt.datetime(2020, 3, 20), 300000, 'March 19th', color = 'orangered')

ax.set(xlabel = "Date", title = "Daily Case count\nFeb - April 2020 for USA")

ax.set_ylabel("Number of Cases", color = "red")

# Formatting

ax.xaxis.set_minor_locator(mdates.DayLocator())

ax.xaxis.set_major_locator(mdates.WeekdayLocator())

ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

plt.savefig('../Images/fig1.US_cases.png')
```

```python
# Generate the plot for Retail and Recreation

fig,ax = plt.subplots()

ax.plot(US_df.date, US_df.cases, color = "red", marker = "o")

ax2 = ax.twinx()

ax2.plot(US_df.date, US_df.Per_decrease_ret_rec, color = 'blue')


# Labels

ax.set(xlabel = "Date", title = "Daily Case count vs Retail & Rec Mobility\nFeb - April 2020 for USA")

ax.set_ylabel("Number of Cases", color = "red")

ax2.set_ylabel("% decrease in traffic retail & recreation", color = "blue")

ax2.set_ylim(-80,80)

# Reference Line

ax2.axhline(y=0, color = 'black', linestyle = '--')

# Formatting

ax.xaxis.set_minor_locator(mdates.DayLocator())

ax.xaxis.set_major_locator(mdates.WeekdayLocator())

ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

plt.show()
```
